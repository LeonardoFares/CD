```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
excel = pd.read_excel('PNAD 2013 A.xlsx')
dados = pd.DataFrame()
```

```python
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
# Definindo as variáveis:
dados['Sexo'] = excel.V0302[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   
dados['Cor_ou_Raça'] = excel.V0404[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   
dados['Rede_de_Ensino'] = excel.V6002[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   
dados['Nivel_de_Ensino'] = excel.V4802[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   
dados['Curso'] = excel.V6003[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   
dados['Idade'] = excel.V8005[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   
dados["Regiao"] = excel.UF[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)] 
dados["Renda_Mensal_Familiar"] = excel.V4750[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   

dados.Sexo = dados.Sexo.astype('category')
dados.Sexo.cat.categories = ('Masculino', 'Feminino')

dados.Cor_ou_Raça = dados.Cor_ou_Raça.astype('category')
dados.Cor_ou_Raça.cat.categories = ('Indígena', 'Branca', 'Preta', 'Amarela', 'Parda')

dados.Rede_de_Ensino = dados.Rede_de_Ensino.astype('category')
dados.Rede_de_Ensino.cat.categories = ('Pública', 'Privada')

dados.Nivel_de_Ensino = dados.Nivel_de_Ensino.astype('category')
dados.Nivel_de_Ensino.cat.categories = ('Ensino Médio', 'Ensino Superior')

dados.Curso = dados.Curso.astype('category')
dados.Curso.cat.categories = ('Regular do Ensino Médio', 'Supletivo do Ensino Médio', 'Superior de Graduação', 'Pré-vestibular')

dados.Idade = dados.Idade.astype('float')

dados.Regiao[(excel.UF >= 10) & (excel.UF < 20)] = 1
dados.Regiao[(excel.UF >= 20) & (excel.UF < 30)] = 2
dados.Regiao[(excel.UF >= 30) & (excel.UF < 40)] = 3
dados.Regiao[(excel.UF >= 40) & (excel.UF < 50)] = 4
dados.Regiao[(excel.UF >= 50) & (excel.UF < 60)] = 5

dados.Regiao = dados.Regiao.astype('category')
dados.Regiao.cat.categories = ('Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste')

dados.Renda_Mensal_Familiar = dados.Renda_Mensal_Familiar.astype('float')

 
dados = dados[(excel.V6003 == 2) | (excel.V6003 == 4) | (excel.V6003 == 5) | (excel.V6003 == 10)]   
dados.head()
```

```python
dados.describe()
```

```python
# Tabelas de Frequências Relativas:
n = 30339

print('Frequências relativas por Sexo:')
fr_sexo = (dados.Sexo.value_counts()/n*100).round(decimals=2)
print(fr_sexo,'\n')

print('Frequências relativas por Cor ou Raça:')
fr_raca_cor = (dados.Cor_ou_Raça.value_counts()/n*100).round(decimals=2)
print(fr_raca_cor,'\n')

print('Frequências relativas por Rede de Ensino:')
fr_rede_ensino = (dados.Rede_de_Ensino.value_counts()/n*100).round(decimals=2)
print(fr_rede_ensino,'\n')

print('Frequências relativas por Nível de Ensino:')
fr_nivel_ensino = (dados.Nivel_de_Ensino.value_counts()/n*100).round(decimals=2)
print(fr_nivel_ensino,'\n')

print('Frequências relativas por Curso:')
fr_curso = (dados.Curso.value_counts()/n*100).round(decimals=2)
print(fr_curso,'\n')

print('Frequências relativas por Região:')
fr_regiao = (dados.Regiao.value_counts()/n*100).round(decimals=2)
print(fr_regiao,'\n')
```

```python
# Gráficos variáveis qualitativas:

plot = fr_sexo.plot(kind='bar',title='Sexo',figsize=(6, 6),color=('b','g','r'))
plt.ylabel('Freq.Relativa (em%)')
plt.xlabel('Sexo')
```
